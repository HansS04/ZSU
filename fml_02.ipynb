## √ökol 1 - Detekce odlehl√Ωch hodnot

```python
df = pd.read_csv('https://raw.githubusercontent.com/rasvob/VSB-FEI-Fundamentals-of-Machine-Learning-Exercises/master/datasets/zsu_cv1_data.csv', sep=',')


# √ökol 1: Detekce odlehl√Ωch hodnot

# Tento k√≥d spr√°vnƒõ implementuje detekci odlehl√Ωch hodnot pomoc√≠ IQR pro SalePrice a GrLivArea,
# a n√°slednƒõ vizualizuje data s rozli≈°en√≠m odlehl√Ωch hodnot.

# V√Ωpoƒçet IQR (mezikvartilov√©ho rozpƒõt√≠) a hranic pro 'SalePrice'
# Q1 (prvn√≠ kvartil) je 25. percentil dat.
Q1_price = df['SalePrice'].quantile(0.25)
# Q3 (t≈ôet√≠ kvartil) je 75. percentil dat.
Q3_price = df['SalePrice'].quantile(0.75)
# IQR je rozd√≠l mezi Q3 a Q1.
IQR_price = Q3_price - Q1_price
# Doln√≠ hranice pro odlehl√© hodnoty je Q1 - 1.5 * IQR.
lower_bound_price = Q1_price - 1.5 * IQR_price
# Horn√≠ hranice pro odlehl√© hodnoty je Q3 + 1.5 * IQR.
upper_bound_price = Q3_price + 1.5 * IQR_price

# V√Ωpoƒçet IQR a hranic pro 'GrLivArea' (Nadzemn√≠ obytn√° plocha v ƒçtvereƒçn√≠ch stop√°ch)
# Q1 pro GrLivArea.
Q1_area = df['GrLivArea'].quantile(0.25)
# Q3 pro GrLivArea.
Q3_area = df['GrLivArea'].quantile(0.75)
# IQR pro GrLivArea.
IQR_area = Q3_area - Q1_area
# Doln√≠ hranice pro odlehl√© hodnoty v GrLivArea.
lower_bound_area = Q1_area - 1.5 * IQR_area
# Horn√≠ hranice pro odlehl√© hodnoty v GrLivArea.
upper_bound_area = Q3_area + 1.5 * IQR_area

# Oznaƒçen√≠ odlehl√Ωch hodnot v nov√©m sloupci 'is_outlier'
# D≈Øm je oznaƒçen jako odlehl√Ω, pokud 'SalePrice' je mimo sv√© hranice NEBO 'GrLivArea' je mimo sv√© hranice.
df['is_outlier'] = ((df['SalePrice'] < lower_bound_price) | (df['SalePrice'] > upper_bound_price)) | \
                   ((df['GrLivArea'] < lower_bound_area) | (df['GrLivArea'] > upper_bound_area))

# Vizualizace odlehl√Ωch hodnot pomoc√≠ scatter plotu
# Vytvo≈ôen√≠ nov√© figury a osy pro plot s definovanou velikost√≠.
plt.figure(figsize=(10, 6))
# Vytvo≈ôen√≠ scatter plotu:
# - data: pou≈æijeme DataFrame 'df'.
# - x: sloupec 'GrLivArea' na ose x.
# - y: sloupec 'SalePrice' na ose y.
# - hue: barva bod≈Ø se bude li≈°it podle hodnoty ve sloupci 'is_outlier' (True/False).
# - alpha: nastaven√≠ pr≈Øhlednosti bod≈Ø pro lep≈°√≠ vizualizaci p≈ôekr√Ωvaj√≠c√≠ch se bod≈Ø.
sns.scatterplot(data=df, x='GrLivArea', y='SalePrice', hue='is_outlier', alpha=0.6)
# Nastaven√≠ n√°zvu grafu.
plt.title('Detekce odlehl√Ωch hodnot v SalePrice a GrLivArea')
# Nastaven√≠ popisku osy x.
plt.xlabel('GrLivArea')
# Nastaven√≠ popisku osy y.
plt.ylabel('SalePrice')
# Zobrazen√≠ grafu.
plt.show()
```
